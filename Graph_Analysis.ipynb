```python
import pandas as pd
import numpy as np
df=pd.read_csv("Data/amazon_reviews.csv")
df.head()
```

```python
df2=pd.DataFrame()
df2["userName"]=df["userName"]
df2["itemName"]=df["itemName"]
df2.head()
```

```python
df2=df2.drop(df2[df2["userName"]=="Amazon Customer"].index)
df2=df2.drop(df2[df2["userName"]=="Kindle Customer"].index)
df2=df2.reset_index()
df2.head()
```

```python
dic2={}
for i in range(len(df2)):
    cur=str(df2["itemName"][i]).replace(","," ")
    if cur not in dic2:
        dic2[cur]=set()
    dic2[cur].add(df2["userName"][i])
#print(dic2)
```

```python
fro=[]
to=[]
for k,v in dic2.items():
    v=list(v)
    for i in range(len(v)):
        p=v[i]
        for j in range(i+1,len(v)):
            q=v[j]
            if p==" ":
                p="NA"
            if q==" ":
                q="NA"
            fro.append(p)
            to.append(q)
len(fro)
len(to)
```

```python
df_node=pd.DataFrame()
df_node["from"]=fro
df_node["to"]=to
df_node["weight"]=[1]*len(df_node)
df_node.head()
```

```python
group=df_node.groupby([df_node["from"], df_node["to"]])
df_group=group.sum()
df_group=df_group.reset_index()
df_group.head()
```

```python
df_group.to_csv("df_group.csv",index=False)
```

```python
df_more=df_group.loc[df_group["weight"]>8]
len(df_more)
df_more.to_csv("df_more.csv",index=False)
```

```python
a = []
f = open('df_more.csv','r',encoding='utf-8')
line = f.readline()
while line:
    a.append(line.replace("\n","").split(","))   
    line = f.readline()
f.close()
```
